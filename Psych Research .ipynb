```python
import pandas as pd
```

```python
sbjNum = [1, 2, 4, 5, 6, 7, 8, 9, 10, 11, 12, 14]
gpData = pd.DataFrame()

for i in sbjNum:
    files = "C:\\Users\\Paklap\\Downloads\\data_old\\SST_s" + str(i) + "_r1.csv" 
    D = pd.read_csv(files)
    D["sbjId"] = i
    gpData = pd.concat([gpData, D], axis=0).reset_index(drop=True)

data = []

df = pd.DataFrame(data, columns = ['P(ChoiceError)', 'Go_sbjRT_Mean', 'Stop_sbjRT_Mean', 'P(Response|StopTrial)',
                                   'P(Response|GoTrial)','SSD_Mean', 'SSRT'], index = ['S1', 'S2', 'S4', 'S5', 'S6', 'S7', 'S8', 'S9', 'S10',
                                                         'S11', 'S12', 'S14'])
```

```python
for i in sbjNum:
    df.loc['S' + str(i), 'P(ChoiceError)'] = gpData.loc[(gpData['sbjId']==i) & (gpData['direction'] != gpData['sbjResp'])].shape[0] / gpData.loc[(gpData['sbjId']==i)].shape[0]
    df.loc['S' + str(i), 'Go_sbjRT_Mean'] = gpData.loc[(gpData['sbjId']==i) & (gpData['trial_type']=='go')].sbjRT.mean()    
    df.loc['S' + str(i), 'Stop_sbjRT_Mean'] = gpData.loc[(gpData['sbjId']==i) & (gpData['trial_type']=='stop')].sbjRT.mean()    
    df.loc['S' + str(i), 'P(Response|StopTrial)'] = gpData.loc[(gpData['sbjId']==i) & (gpData['trial_type']=='stop')].sbjCorr.mean()    
    df.loc['S' + str(i), 'P(Response|GoTrial)'] = gpData.loc[(gpData['sbjId']==i) & (gpData['trial_type']=='go')].sbjCorr.mean()    
    df.loc['S' + str(i), 'SSD_Mean'] = gpData.loc[(gpData['sbjId']==i) & (gpData['trial_type']=='stop')].SSD.mean()
    
    #SSRT
    ranked  = gpData.loc[(gpData['sbjId']==i) & (gpData['trial_type']=='go')].sort_values('sbjRT')
    #ranked = ranked[ranked['sbjRT'].notna()]
    ranked['rank']=ranked['sbjRT'].rank(ascending = True)
    
    percentile_match = round( df.loc['S' + str(i), 'P(Response|GoTrial)'] * ranked.shape[0] )
    df.loc['S' + str(i), 'SSRT'] = percentile_match - df.loc['S' + str(i), 'SSD_Mean']


df['SSRT _Mean'] = df['Stop_sbjRT_Mean'] - df['SSD_Mean']
df
```
